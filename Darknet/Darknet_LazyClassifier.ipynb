```python
from sklearnex import patch_sklearn
patch_sklearn()

import warnings
import numpy as np 
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
from featurewiz import featurewiz
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.svm import LinearSVC
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, balanced_accuracy_score, recall_score, matthews_corrcoef
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectKBest, chi2, f_classif, RFE, SelectPercentile

warnings.filterwarnings("ignore")
dataset =  pd.read_excel(r"C:/Users/earth/Desktop/PracaMagisterska_WitoldWluczkowski/ZbioryDanych/cicdarknet2020lab_clear.xlsx", index_col=0)


def print_top_correlated_features(X_features):
    correlation_matrix = X_features.corr()
    # Indeksy cech w macierzy korelacji
    feature_indices = np.arange(len(correlation_matrix.columns))
    # Kombinacje indeksów cech
    feature_combinations = np.array(np.meshgrid(feature_indices, feature_indices)).T.reshape(-1, 2)
    # Oblicz współczynniki korelacji dla każdej pary cech
    correlation_values = [correlation_matrix.iloc[i, j] for i, j in feature_combinations if i != j]
    # Znajdź 10 największych wartości korelacji
    top_10_correlations = sorted(zip(feature_combinations, correlation_values), key=lambda x: abs(x[1]), reverse=True)[:10]
    # Wypisz cechy i ich współczynniki korelacji
    print("10 najbardziej skorelowanych cech:")
    for (i, j), correlation in top_10_correlations:
        feature_pair = f"{correlation_matrix.columns[i]} - {correlation_matrix.columns[j]}"
        print(f"{feature_pair}: {correlation:.5f}")

def pearson_correlation(X_features):
    # Sprawdź, czy X_features jest typu DataFrame
    if not isinstance(X_features, pd.DataFrame):
        raise ValueError("X_features powinien być obiektem typu DataFrame.")

    # Pobierz listę cech
    features_list = X_features.columns.tolist()

    # Utwórz pustą listę do przechowywania wyników
    correlation_results = []

    # Iteruj po każdej parze cech i oblicz współczynnik korelacji Pearsona
    for i in range(len(features_list)):
        for j in range(i + 1, len(features_list)):
            feature1 = features_list[i]
            feature2 = features_list[j]
            correlation_value = X_features[feature1].corr(X_features[feature2])

            # Dodaj wynik do listy
            correlation_results.append((feature1, feature2, round(correlation_value, 5)))

    # Posortuj listę wyników według wartości współczynnika korelacji
    correlation_results.sort(key=lambda x: abs(x[2]), reverse=True)

    # Wyświetl 10 najlepszych wyników
    print("Top 10 cech i ich współczynniki korelacji:")
    for result in correlation_results[:10]:
        print(f"{result[0]} - {result[1]}: {result[2]}")


def plot_correlation_matrix(features, method_name):
    correlation_matrix = features.corr()
    # Utwórz wykres heatmapy
    plt.figure(figsize=(10, 8))  # Zmniejsz rozmiar wykresu
    heatmap = sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.4, cbar_kws={'label': 'Korelacja'})
    # Dostosuj kolor i rozmiar tekstu na osiach x, y i osi kolorów (cbar)
    heatmap.set_xticklabels(heatmap.get_xticklabels(), fontsize=8, rotation=45, ha='right')
    heatmap.set_yticklabels(heatmap.get_yticklabels(), fontsize=8)
    # Dostosuj kolor i rozmiar tekstu na osi kolorów (cbar)
    cbar = heatmap.collections[0].colorbar
    cbar.ax.tick_params(labelsize=7)
    # Zamiast cbar.set_ylabel, możemy skorzystać z tytułu osi kolorów
    cbar.set_label('Korelacja', fontsize=14)
    plt.title('Macierz Korelacji dla metody: ' + method_name, fontsize=16)
    folder_name='wykresydlaDarknet'  # Stała wartość folder_name
    os.makedirs(folder_name, exist_ok=True)
    plt.savefig(os.path.join(folder_name, f'{method_name}_darknet.png'), transparent=True)
    plt.show()
    
# def get_top_correlated_features(X_features, method_name, threshold=0.3, num_features=10):
#     print()
#     print()
#     """
#     Zwraca listę num_features najbardziej skorelowanych cech na podstawie danej ramki danych X_features.

#     Parameters:
#     - X_features: pandas DataFrame, ramka danych zawierająca cechy
#     - threshold: float, próg korelacji, powyżej którego cechy są uznawane za mocno skorelowane (domyślnie 0.8)
#     - num_features: int, liczba cech do wyboru (domyślnie 6)

#     Returns:
#     - top_correlated_features: list, lista zawierająca num_features najbardziej skorelowanych cech
#     """
#     correlation_matrix = X_features.corr()
#     correlation_pairs = []

#     # Tworzymy pary cech wraz z wartościami bezwzględnymi korelacji
#     for i in range(len(correlation_matrix.columns)):
#         for j in range(i + 1, len(correlation_matrix.columns)):
#             feature_pair = (correlation_matrix.columns[i], correlation_matrix.columns[j], abs(correlation_matrix.iloc[i, j]))
#             correlation_pairs.append(feature_pair)

#     # Sortujemy pary cech według wartości bezwzględnych korelacji w malejącej kolejności
#     correlation_pairs.sort(key=lambda x: x[2], reverse=True)

#     # Wybieramy num_features najbardziej skorelowanych cech
#     top_correlated_features = []
#     i = 0
#     while len(top_correlated_features) < num_features and i < len(correlation_pairs):
#         feature_pair = correlation_pairs[i]
#         if feature_pair[0] not in top_correlated_features and feature_pair[1] not in top_correlated_features:
#             top_correlated_features.append(feature_pair[0])
#             top_correlated_features.append(feature_pair[1])
#         i += 1

#     # Wyświetlamy najbardziej skorelowane cechy w żądanym formacie
#     print(f"{num_features} najbardziej skorelowane cechy dla metody: {method_name}")
#     for feature_pair in correlation_pairs[:num_features]:
#         print(f"{feature_pair[0]} - {feature_pair[1]}: {feature_pair[2]:.5f}")

#     return top_correlated_features[:num_features]

# def get_least_correlated_features(X_features, method_name, num_features=10):
#     print()
#     print()
#     """
#     Zwraca listę num_features najmniej skorelowanych cech na podstawie danej ramki danych X_features.

#     Parameters:
#     - X_features: pandas DataFrame, ramka danych zawierająca cechy
#     - num_features: int, liczba cech do wyboru (domyślnie 10)

#     Returns:
#     - least_correlated_features: list, lista zawierająca num_features najmniej skorelowanych cech
#     """
#     correlation_matrix = X_features.corr()
#     correlation_pairs = []

#     # Tworzymy pary cech wraz z wartościami bezwzględnymi korelacji
#     for i in range(len(correlation_matrix.columns)):
#         for j in range(i + 1, len(correlation_matrix.columns)):
#             feature_pair = (correlation_matrix.columns[i], correlation_matrix.columns[j], abs(correlation_matrix.iloc[i, j]))
#             correlation_pairs.append(feature_pair)

#     # Sortujemy pary cech według wartości bezwzględnych korelacji rosnąco
#     correlation_pairs.sort(key=lambda x: x[2])

#     # Wybieramy num_features najmniej skorelowanych cech
#     least_correlated_features = []
#     i = 0
#     while len(least_correlated_features) < num_features and i < len(correlation_pairs):
#         feature_pair = correlation_pairs[i]
#         if feature_pair[0] not in least_correlated_features and feature_pair[1] not in least_correlated_features:
#             least_correlated_features.append(feature_pair[0])
#             least_correlated_features.append(feature_pair[1])
#         i += 1

#     # Wyświetlamy najmniej skorelowane cechy w żądanym formacie
#     print(f"{num_features} najmniej skorelowane cechy dla metody: {method_name}")
#     for feature_pair in correlation_pairs[:num_features]:
#         print(f"{feature_pair[0]} - {feature_pair[1]}: {feature_pair[2]:.5f}")

#     return least_correlated_features[:num_features]

drop_columns = [ #Kolumny do usunięcia
        "Flow ID",    
        'Fwd Header Length.1',
        "Source IP", "Src IP",
        "Source Port", "Src Port",
        "Destination IP", "Dst IP",
        "Destination Port", "Dst Port",
        "Timestamp", "Label.1"
]
dataset.drop(columns=drop_columns, inplace=True, errors='ignore') # Usuwanie kolumn ze zbioru danych -Dropowanie kolumn, które są Stringami
dataset.drop_duplicates(inplace=True) #Usuwanie duplikatów
dataset.reset_index(inplace=True, drop=True)
dataset = dataset.astype(float, errors='ignore') #Zmiana typów wszystkich kolumn na float ignorując errory
print("This Dataset has {} rows and {} columns".format(dataset.shape[0], dataset.shape[1]))
dataset = dataset.replace((np.inf, -np.inf, np.nan), 0).reset_index(drop=True)

for col in dataset.columns:
    dataset[col][dataset[col] < 0] = 0

# Define the initial features list
all_features = dataset.columns.tolist()[:-1]

X = dataset.iloc[:, :-1].values
X = X.astype('float64')

y = dataset.iloc[:, -1].values
unique_labels = np.unique(y)
mapping = {label: idx for idx, label in enumerate(unique_labels)}
y = np.array([mapping[label] for label in y])

malign = dataset[dataset['Label'] == 1]
benign = dataset[dataset['Label'] == 0]

print('Number of darknet move that has occured:', len(malign), "Percent of occured darknet move:", round((len(malign) / dataset.shape[0]) * 100, 2), '%')
print('Number of non-darknet move:', len(benign), "Percent of non-darknet move", round((len(benign) / dataset.shape[0]) * 100, 2), '%')

X = dataset.drop(['Label'], axis=1)
y = dataset['Label']
from lazypredict.Supervised import LazyClassifier

feature_selection_methods = {
        'RFE': RFE(estimator=RandomForestClassifier(max_depth=5), n_features_to_select=19),
        'SelectKBest(Anova)': SelectKBest(score_func=f_classif, k=19),
        'SelectKBest(chi2)': SelectKBest(score_func=chi2, k = 19),
        'SelectPercentile(Anova)': SelectPercentile(score_func=f_classif, percentile = 25),
        'SelectPercentile(chi2)': SelectPercentile(score_func=chi2, percentile = 25)
    }
#features, train = featurewiz(dataset, target, corr_limit=0.7, verbose=2, sep=",", header=0,test_data="", feature_engg="", category_encoders="")


#print(f"--- Classifier: {clf_name} ---")

for method_name, feature_selector in feature_selection_methods.items():
    #SELEKCJA CECH
    #feature_selector = SelectKBest(chi2, k=i)
    #feature_selector = SelectKBest(chi2, k=5)
    feature_selector.fit_transform(X, y)
    cols_idxs = feature_selector.get_support(indices=True)
    X_features = dataset.iloc[:,cols_idxs]
    pearson_correlation(X_features)
    plot_correlation_matrix(X_features, method_name)
    # get_top_correlated_features(X_features, method_name)
    # get_least_correlated_features(X_features, method_name)
    
    #Podział danych
    X_train, X_test, y_train, y_test = train_test_split(X_features, y, test_size=0.3, shuffle=True)

    #PREPROCESSING 
    
    #SimpleImputer = czyli uzupełnianie brakujących wartości w zbiorze danych.
    imputer = SimpleImputer()
    X_train = imputer.fit_transform(X_train)
    X_test = imputer.transform(X_test)
    
    #PREPROCESSING 
    
    #czyli doprowadzanie danych do podobnej formy, podobnych wartości, aby nie było odchyleń, które mogą przekłamać wyniki
    scaler = MinMaxScaler()
    X_train_std = scaler.fit_transform(X_train)
    X_test_std = scaler.transform(X_test)
    print("---------------------------------------------------------------------------------")
    print("------------------------------METHOD_NAME_IS:",method_name,"------------------")
    print("---------------------------------------------------------------------------------")
    print("Number of features in learning model: ", X_train.shape[1])
    print("Selected features: ", X_features.columns)
    
    pd.set_option("display.precision", 5)
    clf = LazyClassifier(verbose=0,ignore_warnings=True, custom_metric=None)
    models, predictions = clf.fit(X_train_std, X_test_std, y_train, y_test)
    print(models)

target = 'Label'
features, train = featurewiz(dataset, target, corr_limit=0.7, verbose=2, sep=",", header=0, test_data="", feature_engg="", category_encoders="")
print(features)
# plot_correlation_matrix(features, "cos")

X = dataset[features].to_numpy()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=True)

imputer = SimpleImputer()
X_train = imputer.fit_transform(X_train)
X_test = imputer.transform(X_test)
scaler = MinMaxScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)

clf = LazyClassifier(verbose=0, ignore_warnings=True, custom_metric=None)
models, predictions = clf.fit(X_train_std, X_test_std, y_train, y_test)
print(models)
```
